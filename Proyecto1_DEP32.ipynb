```python
##IMPORTACION DE LAS BIBLIOTECAS NECESARIAS

import pyspark

#Creamos una sesion de spark
from pyspark.sql import SparkSession

#CREAMOS UNA INSTANCIA DE SPARK SESSION
spark = SparkSession.builder.appName("Proyecto1").getOrCreate()
```

## Inicio de Consignas

```python
# 1. CREAR UN DATAFRAME CON DATOS DE PERSONAS, QUE INCLUYA NOMBRE, EDAD Y CIUDAD DE RESIDENCIA

#Utilizar el siguiente esquema:
#data = [("Alice", 25, "New York"),
#        ("Bob", 30, "Los Angeles"),
#        ("Charlie", 22, "Chicago")]



#1. SOLUCION:

import random
# Datos iniciales que nos dio la profesora
#data = [
#    ("Alice", 25, "New York"),
#    ("Bob", 30, "Los Angeles"),
#    ("Charlie", 22, "Chicago")
#]

#-> Crearemos mediante codigo y bajo ese formato 15 filas de datos
data=[]
columnas = ["nombres", "edad", "ciudad"]

# Listas de nombres y ciudades
nombres_pool = ["Alice", "Bob", "Charlie", "David", "Emma", "Frank", "Grace", "Henry", "Ivy", "Jack"]
ciudades_pool = ["New York", "Los Angeles", "Chicago", "Houston", "Phoenix", "Philadelphia", "Miami", "Boston"]

# Rellenamos con random
while len(data) < 15:
    nombre = random.choice(nombres_pool)
    edad = random.randint(18, 60)
    ciudad = random.choice(ciudades_pool)
    data.append((nombre, edad, ciudad))

# Crear el DataFrame de PySpark con los 15 datos
df = spark.createDataFrame(data, schema=columnas)
df.show(15)

```

```python
#2. Mostrar solo los nombres de las personas del DataFrame.

#2. SOLUCION:
#Usamos la funcion 'select'
df.select('nombres').show()
```

```python
#3. Filtrar personas cuya edad sea mayor o igual a 25.

#3. SOLUCION (MODO 1)
#usamos select, pero esto nos da tambien datos null cuando no cumple
from pyspark.sql.functions import when
df.select('nombres',when(df.edad>="25","cumple")).show()
```

```python
#3. Filtrar personas cuya edad sea mayor o igual a 25.

#3. SOLUCION (MODO 2)

#El metodo filter al parecer es mas efectivo
from pyspark.sql.functions import filter
df.filter(df.edad >= "25").show()
```

```python
#4. Agregar una nueva columna "Pais" con un valor constante para todas las filas.

#4. SOLUCION

#usamos la funcion lit de las fucniones propias de spark
from pyspark.sql.functions import lit
df_4 = df.withColumn(
    "Pais",
    lit("Peru")
)
df_4.show()
```

```python
#5. Calcular el promedio de edad de todas las personas.

#5. SOLUCION

from pyspark.sql import functions as f#mediante 'f' llamamos a las funciones propias de spark

# Calculamos la edad promedio (ignorando nulos)
edad_promedio = df.select(
    f.mean("edad")
).first()[0]

edad_promedio
```

```python
#6. Ordenar el DataFrame por edad en orden descendente.

#6. SOLUCION

from pyspark.sql.functions import desc

# Ordenamos el DataFrame por la columna "edad" de mayor a menor
df_ordenado = df.orderBy(desc("edad"))

df_ordenado.show()
```

```python
# 7. Agrupar por ciudad y calcular la cantidad de personas en cada ciudad.

#7. SOLUCION

#Usamos la funcion groupBy y agregamos la funcion count
df_por_ciudad = df.groupBy("ciudad").count()

# Muestra el resultado en pantalla
df_por_ciudad.show()
```

```python
# 8. Renombrar la columna "nombres" a "NombreCompleto".

# 8. SOLUCION

# Usamos withColumnRenamed para renombrar la columna
df_renombrado = df.withColumnRenamed("nombres", "NombreCompleto")

# Mostramos el resultado
df_renombrado.show()
```

```python
# 9. Eliminar la columna "Edad" del DataFrame.

# 9. SOLUCION

# Usamos drop para eliminar la columna
df_col_eliminada = df.drop("edad")

#Si quisiera parecer que se borra en el mismo dataframe, usaria un artificio df=df.drop("edad")
# Mostramos el resultado
df_col_eliminada.show()
```

```python
# 10. Realizar una consulta SQL en el DataFrame para seleccionar personas mayores de 20 años.

# 10. SOLUCION

# Creamos una vista temporal para consultarla con Spark SQL
df.createOrReplaceTempView("data_tarea1")
#Ahora hacemos la consulta
spark.sql('''
          SELECT nombres
          FROM data_tarea1
          WHERE edad > 20
          ;
          ''').show()
```

```python
# 11. Calcular la suma total de todas las edades.

# 11. SOLUCION

suma_edades = df.select(
    f.sum("edad")
)#.first()[0] #No he usado lo comentado porque preferi mostrar en minitabla la suma

suma_edades.show()
```

```python
# 12. Calcular la edad mínima y máxima de todas las personas.

# 12. SOLUCION

from pyspark.sql import functions as f

#Usaremos agg y f para calcular el minimo y maximo
min_edad= df.agg(f.min("edad")).first()[0]
max_edad= df.agg(f.max("edad")).first()[0]

print('edad minima = ',min_edad)
print('edad maxima = ',max_edad)

df.filter(df.edad == min_edad).show()
df.filter(df.edad == max_edad).show()
```

```python
# 13. Filtrar personas cuya ciudad de residencia sea "Chicago" y edad sea menor de 30.

# 13. SOLUCION

#Debido a que genere data aleatorio usare una ciudad aleatoria que tenga en el df en lugar de Chicago
# en caso de chicago no se haya generado 
ciudad= "Houston"
df.filter((df.ciudad == ciudad) & (df.edad < 30)).show()
```

```python
# 14. Agregar una nueva columna "EdadDuplicada" que contenga el doble de la edad.

# SOLUCION

df_duplicado = df.withColumn(
    "EdadDuplicada",
    f.col("edad") * 2
)
df_duplicado.show()
```

```python
# 15. Convertir todas las edades en años a meses.

# 15. SOLUCION

df_edad_mes = df.withColumn(
    "edad",
    f.col("edad") *24
)
#Mostrare primero la edad (años) y luego en meses
df_edad_mes=df_edad_mes.withColumnRenamed("edad", "edad_mes")
df.show()
df_edad_mes.show()
```

```python
# 16. Contar el número total de personas en el DataFrame.

# 16. SOLUCION

from pyspark.sql import functions as f

# contamos en la columna de los nombres y extraemos el número escalar final
cant = df.agg(f.count("nombres")).first()[0]

print('cantidad de nombres = ', cant)
```

```python
# 17. Filtrar personas cuya edad sea un número par.

# 17. SOLUCION

#usamos filter
from pyspark.sql.functions import filter
df.filter(df.edad % 2 == 0).show()
```

```python
# 18. Calcular la cantidad de personas por rango de edades (0-20, 21-40, 41-60, 61+).

# 18. SOLUCION

from pyspark.sql import functions as f

#primero clasificaremos los rangos solicitados
df_rangos = df.withColumn(
    "rango_edad",
    f.when((f.col("edad") >= 0) & (f.col("edad") <= 20), "0-20")
     .when((f.col("edad") >= 21) & (f.col("edad") <= 40), "21-40")
     .when((f.col("edad") >= 41) & (f.col("edad") <= 60), "41-60")
     .otherwise("61+")
)

#Ahora groupBy nos dira cuanto se repiten cada rango 
resultado = df_rangos.groupBy("rango_edad").agg(
    f.count("nombres").alias("cantidad_personas")
)

#Finalmente lo ordenamos por rango de menor a mayor
resultado.orderBy("rango_edad").show()
```

```python
# 19. Contar cuántas personas tienen el mismo nombre.

# 19. SOLUCION

from pyspark.sql import functions as f

# Agrupamos por la columna 'nombres' y contamos las repeticiones
df_nombres_repetidos = df.groupBy("nombres").agg(
    f.count("nombres").alias("cantidad")
)

#Ordenamos de menor a mayor y mostramos
df_nombres_repetidos.orderBy(f.asc("cantidad")).show()
```

```python
# 20. Concatenar las columnas "Nombre" y "Ciudad" en una nueva columna llamada "InformacionPersonal".

# 20. SOLUCION

from pyspark.sql import functions as f

# Creamos la nueva columna uniendo 'nombres' y 'ciudad' separados por un guion
df_unido = df.withColumn(
    "InformacionPersonal", 
    f.concat_ws(" - ", f.col("nombres"), f.col("ciudad"))
)

# Mostramos el resultado
df_unido.select("nombres", "ciudad", "InformacionPersonal").show(truncate=False)
```
